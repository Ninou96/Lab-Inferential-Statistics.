# Set up the hypothesis

H0 : There is no diffrence between the sample and population' BP i.e sample' mean is the pop mean.
µ = 120 mmHG

H1 : There is a remarkable difference between the sample and population' BP
µ ≠ 120 mmHG

It tests the null hypothesis that the means are equal.
if the resulting p-value of levene's test is less than some significance level (0,05) then we reject the H0
otherwise we accept it.

```python
from statistics import math
sample_mean = 120
pop_mean = 130.1
pop_std = 21.21
n = 100
z_score = (pop_mean-sample_mean)/(pop_std/math.sqrt(n))
print(f"z_score is {z_score}")
```

```python
data.head(5)
```

```python
numerics = data._get_numeric_data()
```

```python
numerics.head(5)
```

```python
numerics = numerics.drop(['id','time'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

```python
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```

```python
plot_labels = ['PC'+str(i) for i in range(1,len(variations)+1)]
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,5))
plt.bar(x = plot_labels, height=variations)
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA Scree Plot')
```

```python
data = pd.DataFrame(pca.components_)
data = data[[0,1,2]]
data.columns = ['PC1', 'PC2', 'PC3']
data.head()
```
